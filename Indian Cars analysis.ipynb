```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('carinindia.csv')
```

```python
df.head(10)
```

```python
df.shape
```

```python
df.columns
```

```python
#dropping unnecessary columns

df = df[[#'web-scraper-order', 'web-scraper-start-url', 'Links', 'Links-href',
       'Name', 'Price', 'Mileage', 'Engine', 'Safety', 'Fuel Type',
       'Transmission', 'Seating Capacity', #'Pages', 'Pages-href'
]]
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df.columns
```

```python
df = df.rename(columns={'Fuel Type':'Fuel_type','Seating Capacity':'Seating_capacity'})
```

```python
df.head(10)
```

```python
df.info()
```

```python
fuel_type_numbers = df['Fuel_type'].value_counts()
```

```python
plt.figure(figsize=(10,5))
ax = sns.countplot(x='Fuel_type',data=df)
ax.bar_label(ax.containers[0])
plt.title('Number of Cars by Fuel Type')
plt.show()
```

```python
transmission_type_numbers = df['Transmission'].value_counts()
```

```python
c=['r','g','b','m']
plt.figure(figsize=(7,5))
transmission_type_numbers.plot(kind='bar',color=c,alpha=0.5)
plt.title('Number of Cars by Transmission Type')
plt.xlabel('Transmission Type')
plt.ylabel('Count')
plt.show()
```

```python
#\d+: This matches one or more digits.
#\.?: This matches an optional dot (.).
#\d*: This matches zero or more digits.

df['Mileage'] = df['Mileage'].str.extract('(\d+\.?\d*)').astype(float)
```

```python
plt.hist(df['Mileage'],bins=20,edgecolor='black')
plt.title('Distribution of Car Mileage')
plt.xlabel('Mileage(kmpl)')
plt.ylabel('Frequency')
plt.show()
```

```python
df['Engine'] = df['Engine'].str.extract('(\d+\.?\d*)').astype(float)
```

```python
plt.hist(df['Engine'], bins=20, color='salmon', edgecolor='black')
plt.title('Distribution of Car Engine Specifications')
plt.xlabel('Engine (cc)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(x='Price',y='Engine',data=df)
plt.xticks(rotation=90)
plt.title('Engine and Price comparison')
plt.show()
```

```python
plt.figure(figsize=(13,6))
ax = sns.countplot(x=df['Engine'])
ax.bar_label(ax.containers[0])
plt.title('Number of Engine options present in car according to cc')
plt.show()
```

```python
df_best_safety = df[df['Safety'] == '5 Star (Global NCAP)']
```

```python
df_best_safety
```

```python
plt.figure(figsize=(13,5))
sns.countplot(x=df_best_safety['Name'])
plt.title("Cars with 5 star rating")
plt.show()
```

### We observed that German Cars are more in numbers when it comes to safety

```python
df_best_mileage = df[df['Mileage']>=20]
```

```python
plt.figure(figsize=(13,5))
sns.countplot(x=df_best_mileage['Mileage'],hue=df_best_mileage['Name'])
plt.show()
```

### We conclude that Maruti Cars are more in numbers when it comes to Mileage

```python
df.head()
```

```python
df['Safety'] = df['Safety'].str.extract('(\d+)').astype(float)
```

```python
plt.figure(figsize=(5,5))
sns.barplot(x='Safety',y='Mileage',data=df)
plt.title('Mileage vs Safety Rating')
plt.xlabel('Safety Rating')
plt.ylabel('Mileage')
```

```python
df_top5safety = df[df['Safety']==5.0]
df_top5safety = df_top5safety.head(5)
```

```python
df_top5lowest = df[df['Safety']<=3.0]
df_top5lowest = df_top5lowest.head(5)
df_top5lowest
```

```python
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
sns.barplot(data=df_top5safety,x=df_top5safety['Safety'],y=df_top5safety['Mileage'],hue=df_top5safety['Name'])
plt.title('Safest cars w.r.t. mileage')

plt.subplot(1,2,2)
sns.barplot(data=df_top5lowest,x=df_top5lowest['Safety'],y=df_top5lowest['Mileage'],hue=df_top5lowest['Name'])
plt.title('Cars below 2 star safety rating')

plt.show()
```

# Conclusion

* We conclude that cars that are safe, hence they are heavy and in result they are having low mileage whereas cars that are not safe, they are having more mileage
* Maruti is best for mileage but not for safety
* ~1000 and ~1500 cc engines are more in demand that is why companies are making more cars with these engine
* Demand of petrol cars are more than any other fuel type
